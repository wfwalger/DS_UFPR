##Exercise 3: Compound Interest

Pretend you have just opened a new savings account earning 4 percent interest per year. The interest you earn is paid at the end of the year and added to the balance of the savings account. Write a program that begins by reading the amount of money deposited into the account from the user. Then your program should compute and display the amount in the savings account after 1, 2, and 3 years. Display each amount so that it is rounded to 2 decimal places.

```python
def bamerindus(investimento, taxa):
  investimento = float(investimento)
  taxa = float(taxa)
  rendimento=investimento*taxa+investimento
  return rendimento

inicial=input('Digite o investimento inicial R$: ')
taxar=0.04
primeiro_ano=bamerindus(inicial, taxar)
print('Valor do primeiro ano ser√° R$ {:.2f}' .format(primeiro_ano))

segundo_ano=bamerindus(primeiro_ano,taxar)
print('Valor do segundo ano ser√° R$ {:.2f}' .format(segundo_ano))

terceiro_ano=bamerindus(segundo_ano,taxar)
print('Valor do terceiro ano ser√° R$ {:.2f}' .format(terceiro_ano))
```

##Exercise 5: Temperature Conversion Table

Write a program displaying a temperature conversion table for degrees Celsius and Fahrenheit. The table should include rows for all temperatures between 0 and 100 degrees Celsius that are multiples of 10 degrees Celsius. Include appropriate headings on your columns.

```python
def converte(a,b,c):
  a=int(a)
  b=float(b)
  c=int(c)
  resultado=a*b+c
  return resultado

print('| Celsius |','|Farenheit|')
print('|---------| |---------|')
count = 0
b=1.8
c=32
temp_celsius=0
while count <= 100:
  conversao=converte(temp_celsius,b,c)
  print('|',temp_celsius,'|',conversao,'|')
  temp_celsius=temp_celsius+10
  count += 10
```

##Exercise 6: Caesar Cipher - Challenge

One of the first known examples of encryption was used by Julius Caesar. Caesar needed to provide written instructions to his generals, but he didn‚Äôt want his enemies to learn his plans if the message slipped into their hands. As a result, he developed what later became known as the Caesar Cipher.

The idea behind this cipher is simple (and as a result, it provides no protection against modern code-breaking techniques). Each letter in the original message is shifted by 3 places. As a result, A becomes D, B becomes E, C becomes F, D becomes G, etc. The last three letters in the alphabet are wrapped around the beginning: X becomes A, Y becomes B, and Z becomes C. Non-letter characters are not modified by the cipher.

Write a program that implements a Caesar cipher. Allow the user to supply the message and the shift amount and then display the shifted message. Ensure that your program encodes both uppercase and lowercase letters. Your program should also support negative shift values to be used both to encode and decode messages.

```python
#O programa inicia solicitando ao usuario para digitar o texto, depois pergunta se o usu√°rio deseja codificar ou decodificar a mensagem
#Caso a op√ß√£o seja 'codificar' o algoritmo altera a posi√ß√£o em 3 caracteres, A -> D, B -> E assim por diante, o % tem a fun√ß√£o de manter o c√°lculo sobre entre os caracteres.
#Caso a op√ß√£o seja 'decodificar' o algoritmo faz o oposto, seguindo a mesma l√≥gica.

def codificar(texto):
    texto_codificado = ""
    for caractere in texto:
        # para letras maiusculas
        if caractere.isupper():
            texto_codificado += chr((ord(caractere) - 65 + 3) % 26 + 65)
        # para letras minusculas
        elif caractere.islower():
            texto_codificado += chr((ord(caractere) - 97 + 3) % 26 + 97)
        else:
        # caracteres diferentes das letras do alfabeto
            texto_codificado += caractere
    return texto_codificado

def decodificar(texto):
    texto_decodificado = ""
    for caractere in texto:
         # para letras maiusculas
        if caractere.isupper():
            texto_decodificado += chr((ord(caractere) - 65 - 3) % 26 + 65)
        # para letras minusculas
        elif caractere.islower():
            texto_decodificado += chr((ord(caractere) - 97 - 3) % 26 + 97)
        else:
           # caracteres diferentes das letras do alfabeto
            texto_decodificado += caractere
    return texto_decodificado

texto_original = input("Digite o texto: ")
modo = input("Deseja codificar (c) ou decodificar (d) a mensagem? ").lower()

if modo == 'c':
    texto_codificado = codificar(texto_original)
    print("Texto codificado:", texto_codificado)
elif modo == 'd':
    texto_decodificado = decodificar(texto_original)
    print("Texto decodificado:", texto_decodificado)
else:
    print("Op√ß√£o inv√°lida. Por favor, escolha 'c' para codificar ou 'd' para decodificar.")
```
